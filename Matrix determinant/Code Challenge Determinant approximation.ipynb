```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
diagonal = np.eye(50)
vector = np.linspace(0, 1, num=100)

swap_values = np.zeros((1, 100))
scale = np.zeros((1, 100))
#create a 50x50 singular matrix and add linear dependency


Matrix1 = np.random.randint(100, size=(50, 50))

Matrix1[[-1]] = Matrix1[[0]]

for i in range(len(vector)):    
    shifted = Matrix1 + (diagonal*vector[i])
    value = abs(np.linalg.det(shifted))
    scale[0][i] = i
    swap_values[0][i] = value
```

```python
average = np.average(swap_values)

```

```python
plt.plot(scale,swap_values,"o-")
plt.xlabel("Iteration number")
plt.ylabel("Determinant Value")
plt.legend(["Average"+str(average)])
plt.grid()
plt.show()
```
