```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
#Probando como hacer una matriz simetrica cuando las matrices son cuadradas

Matrix1 = np.random.randint(100, size=(10, 10))
Matrix2 = np.random.randint(100, size=(10, 10))

simmetry1 = (Matrix1+Matrix1.T)/2
simmetry2 = (Matrix2+Matrix2.T)/2

#print(simmetry1)
print("---"*10)
print(np.dot(simmetry1,simmetry2))

```

```python
#Probando como hacer una matriz simetrica cuando las matrices tienen diferentes dimensoines

Matrix1NS = np.random.randint(100, size=(10, 2))
Matrix2NS = np.random.randint(100, size=(10, 2))

simmetryNS1 = np.dot(Matrix1NS,Matrix1NS.T)
simmetryNS2 = np.dot(Matrix1NS.T,Matrix1NS)

print(simmetryNS1)

print(simmetryNS2)
```

```python
n = 52
F = np.zeros((n, n), dtype="complex")
w = np.exp(-2*np.pi*np.complex(0,1)/n)
 
for i in range(n):
    for k in range(n):
        m = (i)*(k) # (j - 1)*(k - 1), but python is 0 indexed
        F[i, k] = w**m
 
fig, ax = plt.subplots(1,3)
ax[0].imshow(np.real(F)); ax[1].imshow(np.imag(F)); ax[2].imshow(np.abs(F))          
 
x = np.random.randn(n, 1)
x1 = np.matmul(F, x)
 
from scipy.fft import fft
x2 = fft(x)
 
fig, ax2 = plt.subplots(1,1)
ax2.plot(range(n), np.abs(x1))
ax2.scatter(range(n), np.abs(x2), s=50, marker='o', facecolors='none', edgecolors='r', label='fft on x')
ax2.scatter(range(n), np.abs(fft(x1)), s=50, marker='o', facecolors='none', edgecolors='g', label='fft on x1')
ax2.legend(loc=1)
plt.show()
```
